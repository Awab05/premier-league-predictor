```python
import pandas as pd
```

```python
data_2024 = pd.read_csv('2024-2025.csv')
data_2023 = pd.read_csv('2023-2024.csv')
data_2022 = pd.read_csv('2022-2023.csv')
data_2021 = pd.read_csv('2021-2022.csv')
data_2020 = pd.read_csv('2020-2021.csv')
data_2019 = pd.read_csv('2019-2020.csv')
data_2018 = pd.read_csv('2018-2019.csv')
data_2017 = pd.read_csv('2017-2018.csv')
data_2016 = pd.read_csv('2016-2017.csv')
data_2015 = pd.read_csv('2015-2016.csv')
data_2014 = pd.read_csv('2014-2015.csv')
data_2013 = pd.read_csv('2013-2014.csv')
data_2012 = pd.read_csv('2012-2013.csv')
```

```python
combined_data = pd.concat([data_2024, data_2023, data_2022, data_2021, data_2020, data_2019, data_2018, data_2017, data_2016, data_2015, data_2014, data_2013, data_2012], ignore_index=True)
```

```python
columns_to_keep = [
    'HomeTeam', 'AwayTeam', 'FTHG', 'FTAG', 'FTR', 
    'HS', 'AS', 'HST', 'AST', 'HC', 'AC', 'HY', 'AY', 'HR', 'AR',
    'Date'
]
```

```python
final_data = combined_data[columns_to_keep]
```

```python
final_data
```

```python
final_data.to_csv('final_dataset.csv', index=False)
```

```python
def standardize_date(date_str):
    try:
        # Attempt to parse with %d/%m/%Y
        return pd.to_datetime(date_str, format='%d/%m/%Y')
    except ValueError:
        # Fallback to %d/%m/%y
        return pd.to_datetime(date_str, format='%d/%m/%y')

final_data.loc[:, 'Date'] = final_data['Date'].apply(standardize_date)

```
